# Метанпромбанк - Сегментация пользователей по потреблению

В распоряжении данные о клиентах банка Метанпромбанк. Банк располагается в Ярославле и областных городах: Ростов Великий и Рыбинск. Заказчик - менеджер продукта, который разрабатывает стратегии по удержанию клиентов.

<a href="https://drive.google.com/file/d/1uMbGaXHqMrGLjrOsvyCjOjwFJz3pYQeO/view?usp=sharing">Презентация</a>


**Цель исследования:**

Сегментировать клиентов по количеству используемых продуктов, выявить характеристики каждого сегмента и дать рекомендации по удержанию клиентов.

**Задачи исследования**

* Провести исследовательский анализ данных.
* Сегментировать пользователей на основе данных о количестве потребляемых продуктов.
* Сформулировать и проверить статистические гипотезы.

**Исходные данные**

- `userid` — идентификатор пользователя,
- `score` — баллы кредитного скоринга,
- `City` — город,
- `Gender` — пол,
- `Age` — возраст,
- `Objects` — количество объектов в собственности,
- `Balance` — баланс на счёте,
- `Products` — количество продуктов, которыми пользуется клиент,
- `CreditCard` — есть ли кредитная карта,
- `Loyalty` — активный клиент,
- `estimated_salary` — заработная плата клиента,
- `Churn` — ушёл или нет.

###  Изучение распределения признаков

Построим гистограммы и боксплот по признакам таблицы.

```python
# список с числовыми переменными.
numeric = ['score', 'age', 'balance', 'estimated_salary'] 


# цикл рисует гистограммы и боксплот по признакам
for column in numeric:
    sns.set_style("darkgrid")
    
    fig,(ax1, ax2) = plt.subplots(1,2,figsize=(15,5), gridspec_kw={'width_ratios': [4, 1]})
    
    sns.histplot(df[column], kde=True, ax=ax1, bins = 15)
    ax1.set_ylabel('Количество клиентов')
    ax1.set_xlabel('{}'.format(column))
    ax1.set_title("Распределение параметра {}".format(column))
    
    sns.boxplot(y=column, data=df, ax=ax2)
    ax2.set_title("Boxplot для параметра {}".format(column))
    ax2.set_ylabel('{}'.format(column))
    
    plt.show()
```

По графикам видим, что у параметров `score`, `age`, `balance` распределение близкое к нормальному. 

Данные по зарплате `estimated_salary` распределены равномерно по всей выборке.

### Корреляционный анализ

Преобразуем категориальные переменные `city` и `gender` к количественным, чтобы они участвовали в корреляционном анализе. Используем преобразование one-hot encoding.

```python
df_corr = df.copy()

df_corr['city_orig'] = df_corr['city']
df_corr = pd.get_dummies(df_corr, columns=['gender','city'])

df_corr.head()
```

По графику видим, что между признаками нет высоких корреляций.
* Наиболее сильно коррелирует  отток  с возрастом клиента - 0,29, и с городом Ростов Великий - 0,17

### Выделение сегментов и портретов клиентов по сегментам

По полученным данным можно разделить клиентов на 3 сегмента:  

1. Клиенты, у которых 1 продукт.
2. Клиенты с 2 продуктами.
3. Клиенты с 3 и более продуктами.

Добавим в таблицу столбец с распределением клиентов по сегментам.

```python
def segment_products(products):
    '''
    Функция возвращает группы клиентов по количеству потребляемых продуктов
    
    '''
    if products == 1:
        return '1п'
    if products == 2:
        return '2п'
    return '3+'
```

## Проверка гипотез

### Гипотеза  различия дохода между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним.

Сформулируем гипотезу: 

**H₀**- Различие дохода у клиентов с двумя продуктами и у клиентов с одним продуктом статистически незначимо.

**H₁**- Различие дохода у клиентов с двумя продуктами и у клиентов с одним продуктом статистически значимо.

 Для проверки гипотезы применим метод Манна—Уитни.

```python
alpha = 0.05

results = st.mannwhitneyu(
    df.query('products == 1')['estimated_salary'], 
    df.query('products == 2')['estimated_salary'], 
    alternative = 'two-sided')

print('p-значение:', results.pvalue)

if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Не удалось отвергнуть нулевую гипотезу**: доходы клиентов, которые пользуются двумя продуктами банка, и теми, которые пользуются одним, не различаются.

### Гипотеза различия оттока между теми клиентами, которые пользуются двумя продуктами банка, и теми, которые пользуются одним

Сформулируем гипотезу: 

**H₀**- Отток клиентов, которые пользуются двумя продуктами банка, и отток тех, которые пользуются одним, равны.

**H₁**- Отток клиентов, которые пользуются двумя продуктами банка, и отток тех, которые пользуются одним, различаются.

Так как отток  величина бинарная   применим z-test. Количество наблюдений достаточно.

```python
alpha = 0.05

results, pvalue = stests.ztest(
    df.query('products == 1')['churn'], 
    x2=df.query('products == 2')['churn'],
    value = 0,
    alternative='two-sided')

print('p-значение:', float(pvalue))

if (pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

**Отвергаем нулевую гипотезу**: отток клиентов, которые пользуются двумя продуктами банка, и отток тех, которые пользуются одним, различаются.
